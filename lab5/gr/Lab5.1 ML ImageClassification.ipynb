Το Fashion MNIST προορίζεται ως υποκατάστατο του κλασικού συνόλου δεδομένων MNIST(εικόνες χειρόγραφων ψηφίων (0, 1, 2, κ.λπ.)) το οποίο χρησιμοποιείται συχνά ως "Hello, World" προγραμμάτων μηχανικής μάθησης.

Κάθε εικόνα αντιστοιχεί σε μία μόνο ετικέτα. 

```python
# TensorFlow and tf.keras
import tensorflow as tf
from tensorflow import keras

# Helper libraries
import numpy as np
import matplotlib.pyplot as plt

print(tf.__version__)
```

```python
fashion_mnist = keras.datasets.fashion_mnist

(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

Δεδομένου ότι τα ονόματα των κατηγοριών δεν περιλαμβάνονται στο σύνολο δεδομένων, τα αποθηκεύουμε στο πίνακα class_names  για χρήση αργότερα, κατά την εκτύπωση των εικόνων.

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat','Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

 Οι τιμές των pixel στην πρώτη εικόνα του συνόλου εκπαίδευσης κυμαίνονται από 0 έως 255

Θα κανονικοποιήσουμε τις τιμές και των δύο συνόλων (εκπαίδευσης και ελέγχου)

```python
train_images = train_images / 255.0

test_images = test_images / 255.0
```

## Δημιουργία μοντέλου

Για να δημιουργήσουμε ένα νευρωνικό δίκτυο, απαιτείται η διαμόρφωση των επιπέδων του μοντέλου και στη συνέχεια η μεταγλώττιση του μοντέλου.



###Ρύθμιση των επιπέδων
Το βασικό δομικό στοιχείο ενός νευρωνικού δικτύου είναι το *επίπεδο (layer)*. 

Τα επίπεδα εξάγουν αναπαραστάσεις από τα δεδομένα που τροφοδοτούνται σε αυτά.

Το μεγαλύτερο μέρος της βαθιάς μάθησης συνίσταται στην αλυσίδα απλών επιπέδων. 

Τα περισσότερα επίπεδα, όπως το `tf.keras.layers.Dense`, έχουν παραμέτρους που μαθαίνονται κατά την εκπαίδευση.

![alt text](https://drive.google.com/uc?id=1rkLj69Vq-bRZsZn5yQSMknmcKkHALz0e)

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10)
])
```

Το πρώτο επίπεδο σε αυτό το δίκτυο,  `tf.keras.layers.Flatten`, μετατρέπει τη μορφή των εικόνων από ένα δισδιάστατο πίνακα (28 x 28 pixel) σε μονοδιάστατο πίνακα (28 * 28 = 784 pixel ). Αυτό το επίπεδο δεν έχει παραμέτρους για μάθηση, μόνο διαμορφώνει τα δεδομένα. 


Έπειτα, το δίκτυο περιέχει μια ακολουθία δύο επιπέδων «tf.keras.layers.Dense». Αυτά είναι πλήρως συνδεδεμένα επίπεδα νευρώνων (πυκνά συνδεδεμένα). 
- Το πρώτο επίπεδο «Dense» έχει 128 κόμβους (ή νευρώνες).
- Το δεύτερο (και τελευταίο) επίπεδο επιστρέφει μια συστοιχία με τιμές πιθανοτήτων με μήκος 10. Κάθε κόμβος περιέχει ένα σκορ που δείχνει σε ποια από τις 10 κλάσεις ανήκει η τρέχουσα εικόνα.

### Μεταγλώττιση μοντέλου
 Προτού το μοντέλο είναι έτοιμο για εκπαίδευση, χρειάζεται μερικές ακόμη ρυθμίσεις. Αυτά προστίθενται κατά τη διάρκεια του βήματος *compile* του μοντέλου: 
 * *Συνάρτηση απώλειας (Loss function)* 
  - Μετρά πόσο ακριβές είναι το μοντέλο κατά τη διάρκεια της εκπαίδευσης. Η ελαχιστοποίηση της τιμής της συνάρτησης  "κατευθύνει" το μοντέλο προς τη σωστή κατεύθυνση. 
 *  *Βελτιστοποιητής (Optimizer)* 
  - Ενημερώνει το μοντέλο με βάση τα δεδομένα που δέχεται και την συνάρτηση απώλειας.
 *  *Μετρική* 
  - Χρησιμοποιείται για την παρακολούθηση των βημάτων εκπαίδευσης και δοκιμών. 

Το ακόλουθο παράδειγμα χρησιμοποιεί την *ακρίβεια*, το κλάσμα των εικόνων που έχουν ταξινομηθεί σωστά στο σύνολο των εικόνων.


```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

## Εκπαίδευση του μοντέλου
Η εκπαίδευση του μοντέλου ενός νευρωνικού δικτύου ακολουθεί τα εξης βήματα:
1. Τα δεδομένα εκπαίδευσης τροφοδοτούνται στο μοντέλο. (εδώ τα train_images και train_labels). 
2. Το μοντέλο μαθαίνει να συσχετίζει εικόνες και ετικέτες.
3. Ζητείται από το μοντέλο να κάνει προβλέψεις σχετικά με ένα σύνολο ελέγχου (test_images).
4. Βεβαιώνεται κατά πόσο οι προβλέψεις του δικτύου ταιριάζουν με τις ετικέτες από τον πίνακα test_labels.


### Τροφοδότηση του μοντέλου
Η μέθοδος model.fit τροφοδοτεί το μοντέλο με τα δεδομένα εκπαίδευσης και τις αντίστοιχες ετικέτες με σκοπό την εκπαίδευση του μοντέλου 

```python
model.fit(train_images, train_labels, epochs=10)
```

Καθώς το μοντέλο εκπαιδεύεται, εμφανίζονται οι μετρήσεις των συναρτήσεων απώλειας και ακρίβειας. Αυτό το μοντέλο φτάνει σε ακρίβεια περίπου 0,91 (ή 91%) στα δεδομένα εκπαίδευσης.

###Αξιολόγηση της ακρίβειας του μοντέλου


```python
test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)

print('\nTest accuracy:', test_acc)
```

Αποδεικνύεται ότι η ακρίβεια στο σύνολο δεδομένων ελέγχου είναι λίγο μικρότερη από την ακρίβεια στο σύνολο δεδομένων εκπαίδευσης. Αυτό το κενό μεταξύ της ακρίβειας της εκπαίδευση και της ακρίβειας των ελέγχων αντιπροσωπεύει την υπερεκπαίδευση - **overfitting**. 

Η υπερεκπαίδευση συμβαίνει όταν ένα μοντέλο μηχανικής μάθησης αποδίδει χειρότερα σε νέες εισόδους από ότι στα δεδομένα εκπαίδευσης. 

Ένα υπερεκπαιδευμένο  μοντέλο "απομνημονεύει" το θόρυβο και τις λεπτομέρειες στο σύνολο δεδομένων εκπαίδευσης σε σημείο που επηρεάζει αρνητικά την απόδοση του μοντέλου στα νέα δεδομένα. 

Για περισσότερες πληροφορίες, ανατρέξτε στα ακόλουθα: 
*   [Demonstrate overfitting](https://www.tensorflow.org/tutorials/keras/overfit_and_underfit#demonstrate_overfitting)
*   [Strategies to prevent overfitting](https://www.tensorflow.org/tutorials/keras/overfit_and_underfit#strategies_to_prevent_overfitting)



### Κάνοντας προβλέψεις

Με το εκπαιδευμένο μοντέλο, μπορούμε να κάνουμε προβλέψεις σε νέες εικόνες. Οι μη κανονικοποιημένες προβλέψεις που δημιουργεί το μοντέλο ταξινόμησης, μεταφέρονται στη συνέχεια στην συνάρτηση softmax η οποία δημιουργεί ένα διάνυσμα (κανονικοποιημένων) πιθανοτήτων με μία τιμή για κάθε πιθανή κλάση. 

```python
probability_model = tf.keras.Sequential([model,tf.keras.layers.Softmax()])
```

```python
predictions = probability_model.predict(test_images)
```

 
Εδώ, το μοντέλο έχει προβλέψει την ετικέτα για κάθε εικόνα στο σύνολο ελέγχου. Αν δούμε την πρώτη πρόβλεψη

```python
predictions[0]
```

Μια πρόβλεψη είναι ένας πίνακας 10 αριθμών. Αντιπροσωπεύουν την «εμπιστοσύνη» (confidence) του μοντέλου ότι η εικόνα αντιστοιχεί σε καθένα από τα 10 διαφορετικά είδη ρούχων. Η ετικέτα που έχει την υψηλότερη τιμή εμπιστοσύνης είναι η 9 (ankle boot)

```python
np.argmax(predictions[0])
```

Θα ελέγξουμε την ετικέτα που έχει η 1η εικόνα από το σύνολο ελέγχου για να δούμε αν έγινε σωστή ταξινόμησή της από το μοντέλο.

```python
test_labels[0]
```

Οι ακόλουθες συναρτήσεις θα μας δείξουν το πλήρες σύνολο των προβλέψεων 10 κλάσεων.

```python
def plot_image(i, predictions_array, true_label, img):
  predictions_array, true_label, img = predictions_array, true_label[i], img[i]
  plt.grid(False)
  plt.xticks([])
  plt.yticks([])

  plt.imshow(img, cmap=plt.cm.binary)

  predicted_label = np.argmax(predictions_array)
  if predicted_label == true_label:
    color = 'blue'
  else:
    color = 'red'

  plt.xlabel("{} {:2.0f}% ({})".format(class_names[predicted_label],
                                100*np.max(predictions_array),
                                class_names[true_label]),
                                color=color)

def plot_value_array(i, predictions_array, true_label):
  predictions_array, true_label = predictions_array, true_label[i]
  plt.grid(False)
  plt.xticks(range(10))
  plt.yticks([])
  thisplot = plt.bar(range(10), predictions_array, color="#777777")
  plt.ylim([0, 1])
  predicted_label = np.argmax(predictions_array)

  thisplot[predicted_label].set_color('red')
  thisplot[true_label].set_color('blue')
```

### Επιβεβαίωση προβλέψεων


Ας δούμε κάποιες εικόνες, τις προβλέψεις του μοντέλου και τον πίνακα προβλέψεων. Οι σωστές ετικέτες πρόβλεψης είναι μπλε και οι λανθασμένες ετικέτες πρόβλεψης είναι κόκκινες. Ο αριθμός δίνει το ποσοστό (στα 100) για την προβλεπόμενη ετικέτα.

```python
i = 0
plt.figure(figsize=(6,3))
plt.subplot(1,2,1)
plot_image(i, predictions[i], test_labels, test_images)
plt.subplot(1,2,2)
plot_value_array(i, predictions[i],  test_labels)
plt.show()
```

```python
i = 12
plt.figure(figsize=(6,3))
plt.subplot(1,2,1)
plot_image(i, predictions[i], test_labels, test_images)
plt.subplot(1,2,2)
plot_value_array(i, predictions[i],  test_labels)
plt.show()
```

Αν δούμε τα αποτελέσματα από διάφορες εικόνες του συνόλου ελέγχου μπορούμε να παρατηρήσουμε ότι το μοντέλο μπορεί να είναι λάθος ακόμα και όταν είναι πολύ σίγουρο.

```python
# Plot the first X test images, their predicted labels, and the true labels.
# Color correct predictions in blue and incorrect predictions in red.
num_rows = 5
num_cols = 3
num_images = num_rows*num_cols
plt.figure(figsize=(2*2*num_cols, 2*num_rows))
for i in range(num_images):
  plt.subplot(num_rows, 2*num_cols, 2*i+1)
  plot_image(i, predictions[i], test_labels, test_images)
  plt.subplot(num_rows, 2*num_cols, 2*i+2)
  plot_value_array(i, predictions[i], test_labels)
plt.tight_layout()
plt.show()
```
