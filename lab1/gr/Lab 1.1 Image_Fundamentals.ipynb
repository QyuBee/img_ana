Pictures are represented in Scikit-Image using Numpy Tables.This allows maximum interoperability with other Python libraries, such as Matplotlib and Scipy.

Let's see how to create a Grayscale image as a two -dimensional table:

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
random_image = np.random.random([500, 500])

plt.imshow(random_image, cmap='gray')
plt.colorbar();
```

Μέσα στο πακέτο scikit-image, υπάρχουν πολλά δείγματα εικόνων που παρέχονται από το data module. Ας υποθέσουμε ότι θέλουμε να φορτώσουμε μια μόνο εικόνα για να εκτελέσουμε μερικά πειράματα. Αντί να χρησιμοποιήσουμε μια εξωτερική εικόνα, μπορούμε απλά να φορτώσουμε μία από τις εικόνες που παρέχονται από την [βιβλιοθήκη data του skimage](https://scikit-image.org/docs/dev/api/skimage.data.html?highlight=bool), όπου μπορούμε να εισάγουμε εικόνες για δοκιμές. 

```python
from skimage import data

coins = data.coins()

print('Type:', type(coins))
print('dtype:', coins.dtype)
print('shape:', coins.shape)

plt.imshow(coins, cmap='gray');
```

Μια έγχρωμη εικόνα είναι ένας τρισδιάστατος πίνακας, όπου η τελευταία διάσταση έχει μέγεθος 3 και αντιπροσωπεύει τα 3 κανάλια της εικόνας (κόκκινο, πράσινο και μπλε).

```python
cat = data.chelsea()
print("Shape:", cat.shape)
print("Values min/max:", cat.min(), cat.max())
plt.imshow(cat);
```

Ουσιαστικά, μπορούμε να αλλάξουμε τις τιμές των πινάκων, π.χ. μπορούμε να φτιάξουμε ένα κόκκινο τετράγωνο ως εξής:

```python
cat[10:110, 10:110, :] = [255, 255, 0]  # [red, green, blue]
plt.imshow(cat);
print("Values min/max:", cat.min(), cat.max())
```

## Τύποι δεδομένων και τιμές εικόνας
 Στη βιβλιογραφία, βρίσκουμε διαφορετικές συμβάσεις για την απεικόνιση των τιμών εικόνας:  

*   0 - 255 όπου το 0 είναι μαύρο, το 255 είναι λευκό 
*   0 - 1 όπου το 0 είναι μαύρο, το 1 είναι λευκό 

Το scikit-image υποστηρίζει και τις δύο συμβάσεις - η επιλογή καθορίζεται από τον τύπο δεδομένων του πίνακα.

*Για τα colormaps του Matplotlib δείτε [εδώ](https://scipy-cookbook.readthedocs.io/items/Matplotlib_Show_colormaps.html)*


```python
linear0 = np.linspace(0, 255, 2500).reshape((50, 50))
linear1 = np.linspace(0, 1, 2500).reshape((50, 50)).astype(np.uint8)

print("Linear0:", linear0.dtype, linear0.min(), linear0.max())
print("Linear1:", linear1.dtype, linear1.min(), linear1.max())

fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(15, 15))
ax0.imshow(linear0, cmap='gray')
ax1.imshow(linear1, cmap='gray');
```

Η βιβλιοθήκη scikit-image έχει σχεδιαστεί με τέτοιο τρόπο ώστε να επιτρέπεται οποιοδήποτε τύπος δεδομένων ως είσοδος, αρκεί η περιοχή τιμών του να είναι σωστή (0-1 για floating point, 0-255 unsigned bytes, 0-65535 unsigned 16-bit integers). Μπορείτε να μετατρέψετε εικόνες μεταξύ διαφορετικών αναπαραστάσεων χρησιμοποιώντας το αρχείο img_as_float, img_as_ubyte, κ.τ.λ.

([Utility Functions](https://scikit-image.org/docs/stable/api/skimage.html)

* img_as_float32: single-precision (32-bit) floating point format ([0, 1])

* img_as_float64: double-precision (64-bit) floating point format ([0, 1])

* img_as_uint: unsigned integer format ([0, 65535])

* img_as_int: signed integer format ([-32768, 32767])

* img_as_ubyte: unsigned byte format ([0, 255])

* img_as_bool:boolean format (True or False)

)

```python
from skimage import img_as_float, img_as_ubyte

image = data.chelsea()
plt.imshow(image);
image_ubyte = img_as_ubyte(image)
image_float = img_as_float(image)

print("type, min, max:", image.dtype, image.min(), image.max())
print("type, min, max:", image_ubyte.dtype, image_ubyte.min(), image_ubyte.max())
print("type, min, max:", image_float.dtype, image_float.min(), image_float.max())
print()
print("max ubyte/255 =", image_ubyte.max()/255.)
```

### Slicing and indexing

Tις έγχρωμες εικόνες τις χειριζόμαστε ως τρισδιάστατους πίνακες όπου το R(red) κανάλι της εικόνας περιέχεται στον πρώτο, το G(green) κανάλι στο δεύτερο και το Β(blue) κανάλι στον τρίτο από τους τρεις πίνακες. Δεδομένου ότι η Python έχει μηδενικό δείκτη, μπορούμε να γράψουμε τα εξής:



```python
from skimage import data
import matplotlib.pyplot as plt

color_image = data.chelsea()

# για να πάρουμε το red channel
red_channel = color_image[:, :, 0]  
# ή 
color_image[..., 0]

# ομοίως και για τα άλλα

green_channel = color_image[:, :, 1]  
# ή 
color_image[..., 1]

blue_channel = color_image[:, :, 2]  
# ή 
color_image[..., 2]


print("Red channel dimension",red_channel.shape)
print("Green channel dimension",green_channel.shape)
print("Blue channel dimension",blue_channel.shape)
print("Color image dimension",color_image.shape)

f, (ax0, ax1,ax2,ax3) = plt.subplots(1, 4, figsize=(20, 15))

ax0.imshow(color_image)
ax0.set_title('Color', fontsize=12)
ax1.imshow(red_channel)
ax1.set_title('red channel', fontsize=12)
ax2.imshow(green_channel)
ax2.set_title('green channel', fontsize=12)
ax3.imshow(blue_channel)
ax3.set_title('blue channel', fontsize=12)
```

### Χρωματικοί χώροι
Υπάρχουν διάφοροι χρωματικοί χώροι εκτός του RGB που χρησιμοποιούμε για να αναπαραστήσουμε μια εικόνα.

Συνήθως αποτελούνται από 3 κανάλια, όπου ένα κανάλι τους  αφορά στην ένταση της εικόνας (π.χ. φωτεινότητα) και τα δύο άλλα κανάλια τους  αντιπροσωπεύουν το χρώμα  (π.χ. απόχρωση και κορεσμό σε HSL και HSV). Μπορούμε, με τη χρήση των κατάλληλων μεθόδων να μετατρέψουμε σε διαφορετική αναπαράσταση χρώματος την RGB εικόνα μας. 
(Δείτε [εδώ](https://scikit-image.org/docs/dev/api/skimage.color.html) τις μεθόδους για μετατροπή σε όλους τους χρωματικούς χώρους.

```python
from skimage import color, data, img_as_float,img_as_ubyte
import matplotlib.pyplot as plt
import numpy as np

color_image = img_as_float(data.chelsea())

YCbCr_image =color.rgb2ycbcr(color_image)

YCbCr_image = YCbCr_image.astype(np.uint8)

print("type, min, max:", color_image.dtype, color_image.min(), color_image.max())

print("type, min, max:", YCbCr_image.dtype, YCbCr_image.min(), YCbCr_image.max())

print('YCbCr shape',YCbCr_image.shape)

f, (ax0,ax1) = plt.subplots(1, 2, figsize=(20, 15))

ax0.imshow(color_image)
ax0.set_title('RGB', fontsize=12)

ax1.imshow(YCbCr_image)
ax1.set_title('YCbCr', fontsize=12)
```

### Ιστογράμμμα εικόνας
Το ιστόγραμμα μιας εικόνας μας δείχνει την κατανομή των τιμών των pixels της.

Από το ιστόγραμμα μπορούν να εξαχθούν σημαντικές πληροφορίες για την εικόνα ανάλογα με την κατανομή των τιμών της.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

color_image = data.chelsea()

f, (ax0,ax1,ax2,ax3) = plt.subplots(1, 4)

ax0.hist(color_image[:,:,0], bins=10)
ax0.set_ylabel('amount of pixel')
ax0.set_xlabel('values of pixel')
ax0.set_title('Red Channel')


ax1.hist(color_image[:,:,1], bins=10)
ax1.set_ylabel('amount of pixel')
ax1.set_xlabel('values of pixel')
ax1.set_title('Green Channel')


ax2.hist(color_image[:,:,2], bins=10)
ax2.set_ylabel('amount of pixel')
ax2.set_xlabel('values of pixel')
ax2.set_title('Blue Channel')

ax3.imshow(color_image)
ax3.set_ylabel('RGB image')
```

Δημιουργία συνάρτησης απεικόνισης εικόνας και του ιστογράμματός της.

```python
def show(img):
    # Display the image.
    fig, (ax1, ax2) = plt.subplots(1, 2,
                                   figsize=(12, 3))

    ax1.imshow(img, cmap=plt.cm.gray)
    ax1.set_axis_off()

    # Display the histogram.
    ax2.hist(img.ravel(), lw=0, bins=256)  
    #imag.ravelQ #Επιστρέφεται ένας πίνακας 1-D, που περιέχει τα στοιχεία της εισόδου.
    ax2.set_xlim(0, img.max())
    ax2.set_yticks([])

    plt.show()
```

Χρήση της συνάρτησης απεικόνισης εικόνας και του ιστογράμματός της.


```python
from skimage import data

img = data.camera()
show(img)
```

### Γεωμετρικοί χωρικοί μετασχηματισμοί [1]
Ένας γεωμετρικός μετασχηματισμός αποτελείται από δύο διαδικασίες

1.   ένα χωρικό μετασχηματισμό συντεταγμένων
2.   ένα υπολογισμό των τιμών της έντασης της εικόνας, χρησιμοποιώντας την μέθοδο της παρεμβολής, η οποία αποδίδει τιμές έντασης στα χωρικώς μετασχηματισμένα στοιχεία 

Ένας από τους πιο συνηθισμένους χωρικούς μετασχηματισμούς είναι ο αφινικός μετασχηματισμός
$\begin{bmatrix}x & y & z\end{bmatrix}=\begin{bmatrix} u & w & l\end{bmatrix}T=\begin{bmatrix} u & w & l\end{bmatrix}\begin{bmatrix}
t_{11} & t_{12} & t_{13} \\
t_{21} & t_{22} & t_{23} \\
t_{31} & t_{32} & t_{33}
\end{bmatrix}$



[1] Ενότητα 2.4.4. (Παρεμβολή εικόνας), 2.6.5 (Αφινικοί μετασχηματισμοί)

##Κλιμάκωση 
$Τ=\begin{bmatrix}
c_x & 0 & 0 \\
0 & c_y & 0 \\
0 & 0 & 1
\end{bmatrix}$

```python
from skimage.transform import rescale

image_rescaled_F = rescale(image, 0.25, anti_aliasing=False)
image_rescaled_T = rescale(image, 0.25, anti_aliasing=True)  # ρυθμό Nyquist fs >2fm

show(image)
print(image.shape)
show(image_rescaled_F)
print(image_rescaled_F.shape)
show(image_rescaled_T)
print(image_rescaled_T.shape)
```

#Περιστροφή 
$Τ=\begin{bmatrix}
cοsθ & sinθ & 0 \\
-sinθ & cοsθ & 0 \\
0 & 0 & 1
\end{bmatrix}$

```python
from skimage.transform import rotate


image_rotated = rotate(image, angle=45)
show(image_rotated)

image_rotated_res = rotate(image, angle=45, resize=True)
show(image_rotated_res)
```
