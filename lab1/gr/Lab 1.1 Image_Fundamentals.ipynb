### Slicing and indexing

T          R(red)      ,  G(green)      (blue)       .    Python   ,     :



```python
from skimage import data
import matplotlib.pyplot as plt

color_image = data.chelsea()

#     red channel
red_channel = color_image[:, :, 0]  
#  
color_image[..., 0]

#     

green_channel = color_image[:, :, 1]  
#  
color_image[..., 1]

blue_channel = color_image[:, :, 2]  
#  
color_image[..., 2]


print("Red channel dimension",red_channel.shape)
print("Green channel dimension",green_channel.shape)
print("Blue channel dimension",blue_channel.shape)
print("Color image dimension",color_image.shape)

f, (ax0, ax1,ax2,ax3) = plt.subplots(1, 4, figsize=(20, 15))

ax0.imshow(color_image)
ax0.set_title('Color', fontsize=12)
ax1.imshow(red_channel)
ax1.set_title('red channel', fontsize=12)
ax2.imshow(green_channel)
ax2.set_title('green channel', fontsize=12)
ax3.imshow(blue_channel)
ax3.set_title('blue channel', fontsize=12)
```

```python
from skimage import color, data, img_as_float,img_as_ubyte
import matplotlib.pyplot as plt
import numpy as np

color_image = img_as_float(data.chelsea())

YCbCr_image =color.rgb2ycbcr(color_image)

YCbCr_image = YCbCr_image.astype(np.uint8)

print("type, min, max:", color_image.dtype, color_image.min(), color_image.max())

print("type, min, max:", YCbCr_image.dtype, YCbCr_image.min(), YCbCr_image.max())

print('YCbCr shape',YCbCr_image.shape)

f, (ax0,ax1) = plt.subplots(1, 2, figsize=(20, 15))

ax0.imshow(color_image)
ax0.set_title('RGB', fontsize=12)

ax1.imshow(YCbCr_image)
ax1.set_title('YCbCr', fontsize=12)
```

```python
def show(img):
    # Display the image.
    fig, (ax1, ax2) = plt.subplots(1, 2,
                                   figsize=(12, 3))

    ax1.imshow(img, cmap=plt.cm.gray)
    ax1.set_axis_off()

    # Display the histogram.
    ax2.hist(img.ravel(), lw=0, bins=256)  
    #imag.ravelQ #   1-D,      .
    ax2.set_xlim(0, img.max())
    ax2.set_yticks([])

    plt.show()
```
